```python
%pylab inline
import nba
```

```python
m12b_orbit = np.loadtxt('../data/orbits/m12/m12b_sat_orbit.txt')
```

```python
t =  m12b_orbit[:,0]
pos = m12b_orbit[:,1:4]
vel = m12b_orbit[:,4:7]
```

```python
plt.plot(t, norm(pos, axis=1))
```

```python
satkin = nba.kinematics.Kinematics(pos[0:200], vel[0:200])
```

```python
l, b = satkin.orbpole()
```

```python
fig = plt.subplot(111,  projection="mollweide")
plt.scatter(l*np.pi/180.,b*np.pi/180.)
```
