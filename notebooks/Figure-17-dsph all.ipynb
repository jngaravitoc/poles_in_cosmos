```python
dsph_m12c = np.loadtxt('../data/delta_sph/dshp_m12c.txt')
dsph_m12f = np.loadtxt('../data/delta_sph/dshp_m12f.txt')
dsph_m12r = np.loadtxt('../data/delta_sph/dshp_m12r.txt')
dsph_m12w = np.loadtxt('../data/delta_sph/dshp_m12w.txt')
dsph_m12m =np.loadtxt('../data/delta_sph/dshp_m12m.txt')
```

```python
def host_times(sim):
    snap_times = "/mnt/ceph/users/firesims/fire2/metaldiff/{}_res7100/snapshot_times.txt".format(sim)
    times = np.loadtxt(snap_times, usecols=3)
    return times[:] 
```

```python
times = host_times('m12b')
```

```python
colors = plt.cm.inferno(np.linspace(0,1,300))
sm = plt.cm.ScalarMappable(cmap=plt.cm.inferno, norm=plt.Normalize(vmin=times[300], vmax=times[600]))

fig , ax = plt.subplots(1,5, figsize=(18, 4), sharey=True, sharex=True)
ax[0].grid(alpha=0.3)
ax[1].grid(alpha=0.3)
ax[2].grid(alpha=0.3)
ax[3].grid(alpha=0.3)
ax[4].grid(alpha=0.3)

for i in range(0, 200):
    ax[0].plot(np.arange(3, 12), dsph_m12c[i], c=colors[i+100], alpha=0.4, lw=0.2)

    #ax[1].plot(np.arange(3, 12), dsph_m12c[-i], c=colors[i], alpha=0.4, lw=0.2)
 
for i in range(0,150):
    ax[1].plot(np.arange(3, 12), dsph_m12f[i], c=colors[i+80], alpha=0.4, lw=0.2)
    
for i in range(0,146):
    ax[2].plot(np.arange(3, 12), dsph_m12r[i], c=colors[i+80], alpha=0.4, lw=0.2)
    
for i in range(0, 300):
    ax[3].plot(np.arange(3, 12), dsph_m12w[i], c=colors[i], alpha=0.4, lw=0.2)

for i in range(0,249):
    ax[4].plot(np.arange(3, 12), dsph_m12m[i], c=colors[i], alpha=0.4, lw=0.2)
    
    

##  Pericenters ---

ax[0].plot(np.arange(3, 12), dsph_m12c[151], c=colors[151], lw=3, ls='--')
#ax[1].plot(np.arange(3, 12), dsph_m12c[151], c=colors[151], lw=2, ls='--')
#ax[2].plot(np.arange(3, 12), dsph_m12f[27], c=colors[27], lw=3, ls='--')
ax[1].plot(np.arange(3, 12), dsph_m12f[83], c=colors[83], lw=3, ls='--')

ax[2].plot(np.arange(3, 12), dsph_m12r[99], c=colors[99], lw=3, ls='--')
#ax[3].plot(np.arange(3, 12), dsph_m12r[262], c=colors[262], lw=2, ls='--')
#ax[3].plot(np.arange(3, 12), dsph_m12r[214], c=colors[214], lw=2, ls='--')

#ax[4].plot(np.arange(3, 12), dsph_m12w[59], c=colors[59], lw=2, ls='--')
ax[3].plot(np.arange(3, 12), dsph_m12w[191], c=colors[191], lw=3, ls='--')
#ax[4].plot(np.arange(3, 12), dsph_m12w[18], c=colors[18], lw=2, ls='--')
    
cax = plt.axes([0.92, 0.1, 0.02, 0.8])
cbar = fig.colorbar(sm, cax=cax)
cbar.set_label(r'$t_{cosmic}\rm{[Gyr]}$')
ax[0].set_xlim(2.5, 11.5)
ax[0].set_ylim(0, 110)

#ax.plot(np.arange(3, 12), dsph[-65], c='k', alpha=0.6, lw=1, ls='--')
#ax.plot(np.arange(3, 12), dsph[149], c='k', alpha=0.6, lw=1, ls='--')
ax[0].set_xlabel('$k$')
ax[1].set_xlabel('$k$')
ax[2].set_xlabel('$k$')
ax[3].set_xlabel('$k$')
ax[4].set_xlabel('$k$')

ax[0].set_ylabel(r'$\rm{\Delta_{sph}(k)}$')

#ax[1][0].set_xlabel('k')
ax[0].set_title('$m12c$', fontsize=22)
ax[1].set_title('$m12f$', fontsize=22)
ax[2].set_title('$m12r$', fontsize=22)
ax[3].set_title('$m12w$', fontsize=22)
ax[4].set_title('$m12m$', fontsize=22)
ax[0].set_xticks([3, 5, 7, 9, 11])

plt.savefig('../plots/final_paper_plots/dsph_all_latte.pdf', bbox_inches='tight')
```
