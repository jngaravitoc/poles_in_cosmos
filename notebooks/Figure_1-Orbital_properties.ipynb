```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('~/matplotlib.mplstyle')

#import nba
import sys
import numpy as np
from matplotlib import colors
from scipy.linalg import norm
```

```python
data = np.loadtxt('latte_satellite_orbit_properties.txt')
```

```python
e_peri = data[:,0]
r_peri = data[:,1]
sat_mass = data[:,5]
host_mass = data[:,6]
```

```python
# From 
host_mass2 = [60, 50, 67, 0, 72, 80, 25, 30, 27, 30, 36, 32]
```

```python
LMC_peris = [42.8398465246102, 42.736377917126234, 44.3313022980394, 42.60247644363717]
e_LMC = [0.8261510434811363, 0.8069443576415316, 0.7425078579718087, 0.7438234561620495]
```

```python
# Color bar should be Mass ratios 
plt.figure(figsize=(6,4))
cm = plt.cm.get_cmap('Reds')
minv = 0.0
maxv = 0.45


plt.scatter(r_peri[1], e_peri[1], label=r'$m12c$',
            c=sat_mass[1]/host_mass[1], marker='D', 
            s=70, cmap=cm, vmin=minv, vmax=maxv,edgecolors='k')

plt.scatter(r_peri[2], e_peri[2], label=r'$m12f$',
            c=sat_mass[2]/host_mass[2], marker='s', 
            s=70, cmap=cm, vmin=minv, vmax=maxv,edgecolors='k')

plt.scatter(r_peri[3], e_peri[3],c=sat_mass[3]/host_mass[3], marker='s',
            s=70, cmap=cm, vmin=minv, vmax=maxv,edgecolors='k')

plt.scatter(r_peri[4], e_peri[4], label=r'$m12i$',
            c=sat_mass[4]/host_mass[4], marker='<', 
            s=70, cmap=cm, vmin=minv, vmax=maxv,edgecolors='green', lw=1)

plt.scatter(r_peri[5], e_peri[5], label=r'$m12m$',
            c=sat_mass[5]/host_mass[5], marker='>', s=70, 
            cmap=cm, vmin=minv, vmax=maxv,edgecolors='k')

plt.scatter(r_peri[6], e_peri[6], label=r'$m12r$',
            c=sat_mass[6]/host_mass[6], marker='p', s=70, 
            cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')


plt.scatter(r_peri[7], e_peri[7], c=sat_mass[7]/host_mass[7], 
            marker='p', s=70, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')


plt.scatter(r_peri[8], e_peri[8], c=sat_mass[8]/host_mass[8], 
            marker='p', s=70, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')



plt.scatter(r_peri[9], e_peri[9], label=r'$m12w$',c=sat_mass[9]/host_mass[9], 
            marker='h', s=70, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')




plt.scatter(r_peri[10], e_peri[10], c=sat_mass[10]/host_mass[10], marker='h',
            s=70, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')




plt.scatter(r_peri[11], e_peri[11], c=sat_mass[11]/host_mass[11], marker='h',
            s=70, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')


plt.scatter(r_peri[0], e_peri[0], label=r'$m12b$', 
            c=sat_mass[0]/host_mass[0], marker='^',
            s=70, cmap=cm, vmin=minv, vmax=maxv,edgecolors='green', lw=1)


sc = plt.scatter(LMC_peris, e_LMC, c=[0.8/12, 1/12, 1.8/12, 2.5/12], label=r'$\rm{LMC}$', marker='*',
                 s=160, cmap=cm, vmin=minv, vmax=maxv, edgecolors='k')

cbar = plt.colorbar(sc)
cbar.set_label(r'$\rm{Host-Sat.\ mass\ ratio}$')
#plt.xlim(0, 90)
plt.ylim(0.3, 1.05)

plt.legend(fontsize=13, ncol=2)
plt.xlabel(r'$r_{peri}\ \rm{[kpc]}$', fontsize=18)
plt.ylabel(r'$\rm{Eccentricity}$', fontsize=18)
#plt.savefig('fire_e_rperi.pdf', bbox_inches='tight')
plt.grid(alpha=0.5)
#plt.show()
plt.savefig('fire_e_rperi.pdf', bbox_inches='tight')
#plt.show()
```
